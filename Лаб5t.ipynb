```
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
from google.colab import files
uploaded = files.upload()
```

```
df=pd.read_csv('Future50.csv')
df.head()
```

```
df.info()
```

```
df.drop(['Rank', 'Franchising', 'Location'], axis=1, inplace=True)
```

```
df.head()
```

```
df.info()
```

```
df['YOY_Sales']=df['YOY_Sales'].str.replace(r"%", "")
df['YOY_Units']=df['YOY_Units'].str.replace(r"%", "")
```

```
df['YOY_Sales']= df['YOY_Sales'].astype('float64')
df['YOY_Units']= df['YOY_Units'].astype('float64')
```

```
df.head()
```

```
df.info()
```

Визуализация

```
df.head()
```

```
sns.boxenplot(x = df['YOY_Sales'])
```

```
sns.boxplot(x='YOY_Sales',data=df)
```

```
sns.kdeplot(df['YOY_Sales'])
```

```
df1 = df.groupby('Unit_Volume')['YOY_Sales'].mean()
```
