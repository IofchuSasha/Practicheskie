```
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
from google.colab import files
uploaded = files.upload()
```

```
df=pd.read_csv('Future50(1).csv')
df.head()
```

```
df.info()
```

```
df.drop(['Rank','Location','YOY_Sales','YOY_Units'], axis=1, inplace=True)
```

```
df.head()
```

```
df.info()
```

Визуализация

```
df.head()
```

```
sns.boxenplot(x = df['Unit_Volume'])
plt.title('boxplot')
```

```
sns.boxplot(x='Sales',data=df)
plt.title('boxplot')
```

```
sns.kdeplot(df['Unit_Volume'])
plt.title('kdeplot')
```

```
df1 = df.groupby('Unit_Volume')['Sales'].mean()
```

```
sns.violinplot(x = df1)
plt.title('subplot')
```

```
sns.heatmap(df.corr(),annot=True,cmap='YlGnBu')
plt.title('heatmapplot')
```

```
sns.regplot(x='Sales', y='Unit_Volume', data=df)
plt.title('regplot')
```

```
sns.pairplot(data=df)
plt.title('pairplot')
```

```
sns.pairplot(df, hue='Franchising', palette='Set1')
```

```
sns.jointplot(x='Sales', y='Unit_Volume', data=df[:50])
plt.title('jointplot')
```
